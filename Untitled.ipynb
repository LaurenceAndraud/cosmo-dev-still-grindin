```python
image_folder = 'images/'
```

```python
# Charger toutes les images dans le dossier et les prétraiter pour le Transfer Learning
image_list = os.listdir(image_folder)
features = np.zeros((len(image_list), 4096))
for i, image_name in enumerate(image_list):
    img = Image.open(os.path.join(image_folder, image_name))
    img = ImageOps.grayscale(img)  # Conversion en niveaux de gris
    img = ImageOps.equalize(img)   # Égalisation de l'histogramme des couleurs
    img.save(os.path.join(image_folder, image_name))
```

```python
# Générer des descripteurs "bag-of-images" à partir des features extraits
kmeans = KMeans(n_clusters=3, random_state=0).fit(features)
boi = np.zeros((3,))
for label in kmeans.labels_:
    boi[label] += 1
```
